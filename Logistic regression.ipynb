```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

dataset = load_iris()
X = dataset.data[:, :2]
y = (dataset.target != 0) * 1
y = y.reshape((150, 1))
X.shape
params = np.zeros((3, 1))
```

```
def sigmoid(z):
  return 1/(1 + np.exp(-z))

```

```
def loss_func(X, y, params):
  z = np.dot(X, params)
  h = sigmoid(z)
  return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
```

```
def gradient_descent(X, y, params, lr, n_iter):
  History = np.zeros((n_iter, ))
  output = np.zeros((n_iter, 150, 1))
  for i in range(n_iter):
    z = np.dot(X, params)
    h = sigmoid(z)
    params = params - (lr * np.dot(X.T, (h - y)))
    History[i] = loss_func(X, y, params)
    output[i] = h
  return (History, params, output)
```

```
p = np.ones((150, 1))
X = np.hstack((p , X))
```

```
n_iter = 20000
lr = 0.0001

initial_loss = loss_func(X, y, params)
print("initial loss = ", initial_loss)

(history, par, output) = gradient_descent(X, y, params, lr, n_iter)

print("last cost: ", history[-1])

plt.plot(range(n_iter), history, 'r')

plt.show()
```
